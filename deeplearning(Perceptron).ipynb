##Perceptron (Binary Classification)
###Example:study hours -> pass(1)/Fail(0)

```python
import numpy as np
```

##Prepare Training Data

```python
#training Data
x=np.array([1,2,3,4]) #Study Hours
y=np.array([0,0,1,1]) # Fail(0)/Pass(1)
x,y
```

## Initialize Parameters

```python
w=0.0
b=0.0
learning_rate=1
epochs=5
print("Initial Weight:",w)
print("Initial Bias:",b)
```

## Define Activation Function(Step Function)

```python
def step_function(z):
  return 1 if z>=0 else 0
```

##Training Loop

```python
for epoch in range(epochs):
  print(f"\nEpoch {epoch+1}")
  for i in range(len(x)):
    #1.Compute weighted sum
    z=w * x[i] + b
    #2.Apply Activation function
    y_pred=step_function(z)
    #3.compute the erro
    error=y[i] - y_pred
    #4.update bias
    w=w+learning_rate*error*x[i]
    #5.update Bias
    b=b+learning_rate*error
    print(f"input:{x[i]},predicted:{y_pred},Actual:{y[i]},w:{w},b:{b}")
```

```python
print("\n Final weight:",w)
print("final bias:",b)
```

##Testing the Model

```python
test_value=2.9
z=w*test_value+b
prediction=step_function(z)
print(f"for study hours {test_value},prediction:",prediction)
```

##Perceptron for OR Gate


```python
import numpy as np

# Input dataset
x = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

# OR gate outputs
y = np.array([0,1,1,1])

# Step function
def step_function(z):
    if z >= 0:
        return 1
    else:
        return 0

# Initialize weights and bias
w = np.zeros(2)
b = 0
learning_rate = 0.1
epochs = 10

print("Initial weights:", w)
print("Initial bias:", b)

# Training
for epoch in range(epochs):
    print(f"\nEpoch {epoch+1}")
    for i in range(len(x)):
        z = np.dot(w, x[i]) + b
        y_pred = step_function(z)
        error = y[i] - y_pred

        w = w + learning_rate * error * x[i]
        b = b + learning_rate * error

        print(f"Input:{x[i]}, Predicted:{y_pred}, Actual:{y[i]}, w:{w}, b:{b}")

print("\nFinal Weights:", w)
print("Final Bias:", b)

# Testing
print("\nTesting OR Gate")
for i in range(len(x)):
    z = np.dot(w, x[i]) + b
    prediction = step_function(z)
    print("Input:", x[i], "Prediction:", prediction)
```
